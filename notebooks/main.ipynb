```python
import sys
import os

# Adding project directory to sys.path
sys.path.append(os.path.abspath('../src'))

from dataset import load_data
from features import preprocess_data, split_data
from modeling.train import train_model
from modeling.predict import predict_user_behavior
from plots import plot_data_distributions, plot_feature_importance, plot_confusion_matrix

# Load and preprocess data
data = load_data('../data/raw/user_behavior_dataset.csv')
data = preprocess_data(data)

# Visualize data distributions
plot_data_distributions(data)


```

```python
# Split data into features and target
X_train, X_test, y_train, y_test, features, scaler = split_data(data)

# Train the model and get the best model
best_model = train_model(X_train, y_train)

# Evaluate the model on the test data and plot the Confusion Matrix
y_pred = best_model.predict(X_test)  # Get predictions
cm = plot_confusion_matrix(best_model, X_test, y_test)  # Plot confusion matrix

# Visualize feature importance
plot_feature_importance(best_model, features)
```
